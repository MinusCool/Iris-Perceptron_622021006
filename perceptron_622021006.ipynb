```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('Percobaan3_622021006.npz')

setosa = data['setosa']
versicolor = data['versicolor']
```

```python
X_setosa = setosa[:, :2]
X_versicolor = versicolor[:, :2]

y_setosa = np.zeros(len(X_setosa))
y_versicolor = np.ones(len(X_versicolor))
```

```python
X = np.vstack((X_setosa, X_versicolor))
y = np.hstack((y_setosa, y_versicolor))
```

```python
w = np.zeros(X.shape[1])
b = 0
learning_rate = 0.1
epochs = 20
```

```python
def activation(x):
    return 1 if x >= 0 else 0
```

```python
for epoch in range(epochs):
    for i in range(len(X)):
        y_pred = activation(np.dot(w, X[i]) + b)
        error = y[i] - y_pred
        w += learning_rate * error * X[i]
        b += learning_rate * error

print("Bobot akhir:", w)
print("Bias akhir:", b)
```

```python
for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i][0], X[i][1], color='red', label='Setosa' if i == 0 else "")
    else:
        plt.scatter(X[i][0], X[i][1], color='blue', label='Versicolor' if i == 50 else "")
```

```python
x_line = np.linspace(min(X[:, 0]), max(X[:, 0]), 100)
y_line = -(w[0] * x_line + b) / w[1]
plt.plot(x_line, y_line, color='black', label='Decision Boundary')

plt.xlabel("Fitur 1")
plt.ylabel("Fitur 2")
plt.title("Perceptron Frank Rosenblatt")
plt.legend()
plt.grid(True)
plt.show()
```
